```python
def permute(input_bits, input_table):
    result = []
    for position in input_table:
        result.append(input_bits[position-1])
    return result

def Lshift(bits, num_positions):
    shifted = []
    for i in range(num_positions, len(bits)):
        shifted.append(bits[i])
    for i in range(num_positions):
        shifted.append(bits[i])
    return shifted

def xor(bits_one, bits_two):
    output = []
    for i in range(len(bits_one)):
        output.append(bits_one[i] ^ bits_two[i])
    return output
```

```python
def f_function(R, K):
    #Expand R from 32 to 48 bits
    expanded_R = permute(R, E_bit_selection_table)

    #XOR with round key
    xored = xor(expanded_R, K)

    result = []

    #Split into 8 groups for S-boxes
    for i in range(8):
        part = xored[i*6:(i+1)*6]

        #Get row from first and last bit
        row = int("".join(map(str, [part[0], part[5]])), 2)

        #Get column from middle bits
        col = int("".join(map(str, part[1:5])), 2)

        output = Sbox_X[i][row][col]

        #Convert output to 4 bits
        for j in range(4):
            bit = (output >> (3 - j)) & 1
            result.append(bit)

    #Apply permutation P
    f = permute(result, P)

    return f
```

```python
def generate_16_round_keys(keyBits):
    #Apply PC1
    keyBits56 = permute(keyBits, PC_1)

    #Split into two halves
    C = keyBits56[:28]
    D = keyBits56[28:]

    roundKeys = []

    for i in range(16):
        #Shift halves
        C = Lshift(C, LeftShifts[i])
        D = Lshift(D, LeftShifts[i])

        #Apply PC2
        Kn = permute(C + D, PC_2)
        roundKeys.append(Kn)

    return roundKeys
```

```python
def des_encryption(plainBits, roundKeys):
    #Apply IP
    ip_result = permute(plainBits, IP)

    #Split into L and R
    L = ip_result[:32]
    R = ip_result[32:]

    for n in range(16):
        K = roundKeys[n]

        #Apply f function
        f = f_function(R, K)

        #Swap and XOR
        new_L = R
        new_R = xor(L, f)

        L = new_L
        R = new_R

    #Final swap
    swap = R + L

    #Apply inverse IP
    cipherBits = permute(swap, inverseIP)

    return cipherBits
```

```python
def des_decryption(cipherBits, roundKeys):
    #Apply IP
    ip_result = permute(cipherBits, IP)

    #Split into L and R
    L = ip_result[:32]
    R = ip_result[32:]

    for n in range(16):
        K = roundKeys[15 - n]

        #Apply f function
        f = f_function(R, K)

        #Swap and XOR
        new_L = R
        new_R = xor(L, f)

        L = new_L
        R = new_R

    #Final swap
    swap = R + L

    #Apply inverse IP
    plainBits = permute(swap, inverseIP)

    return plainBits
```

```python
def triple_des_encrypt_block(blockBits, roundKeys1, roundKeys2, roundKeys3):
    #Encrypt with key1
    step1 = des_encryption(blockBits, roundKeys1)

    #Decrypt with key2
    step2 = des_decryption(step1, roundKeys2)

    #Encrypt with key3
    step3 = des_encryption(step2, roundKeys3)

    return step3


def triple_des_decrypt_block(blockBits, roundKeys1, roundKeys2, roundKeys3):
    #Decrypt with key3
    step1 = des_decryption(blockBits, roundKeys3)

    #Encrypt with key2
    step2 = des_encryption(step1, roundKeys2)

    #Decrypt with key1
    step3 = des_decryption(step2, roundKeys1)

    return step3
```

```python
def string_to_bits(text):
    bits = []

    #Convert each character to bits
    for byte in text.encode("utf-8"):
        for b in format(byte, "08b"):
            bits.append(int(b))

    return bits


def bits_to_string(bits):
    output = ""

    #Convert bits back to characters
    for i in range(0, len(bits), 8):
        byte = bits[i:i+8]
        output += chr(int("".join(map(str, byte)), 2))

    return output


def split_blocks(bits):
    blocks = []

    #Split into 64-bit blocks
    for i in range(0, len(bits), 64):
        blocks.append(bits[i:i+64])

    return blocks
```

```python
def triple_des_encrypt_message(plaintext, key1, key2, key3):
    #Convert plaintext to bits
    plainBits = string_to_bits(plaintext)

    #Split into blocks
    blocks = split_blocks(plainBits)

    #Convert keys
    keyBits1 = string_to_bits(key1)
    keyBits2 = string_to_bits(key2)
    keyBits3 = string_to_bits(key3)

    #Generate round keys
    roundKeys1 = generate_16_round_keys(keyBits1)
    roundKeys2 = generate_16_round_keys(keyBits2)
    roundKeys3 = generate_16_round_keys(keyBits3)

    cipherBits = []

    for block in blocks:
        #Pad if needed
        while len(block) < 64:
            block.append(0)

        encrypted_block = triple_des_encrypt_block(block, roundKeys1, roundKeys2, roundKeys3)
        cipherBits.extend(encrypted_block)

    return cipherBits
```

```python

def triple_des_decrypt_message(cipherBits, key1, key2, key3):
    blocks = split_blocks(cipherBits)

    keyBits1 = string_to_bits(key1)
    keyBits2 = string_to_bits(key2)
    keyBits3 = string_to_bits(key3)

    roundKeys1 = generate_16_round_keys(keyBits1)
    roundKeys2 = generate_16_round_keys(keyBits2)
    roundKeys3 = generate_16_round_keys(keyBits3)

    plainBits = []

    for block in blocks:
        decrypted_block = triple_des_decrypt_block(block, roundKeys1, roundKeys2, roundKeys3)
        plainBits.extend(decrypted_block)

    return bits_to_string(plainBits)
```
