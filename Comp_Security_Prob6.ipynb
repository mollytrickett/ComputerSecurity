```python
#rearranges using different tables
def permute(input_bits, input_table):
    result = []
    for position in input_table:
        result.append(input_bits[position-1])
    return result

#rotates bits to the left
def Lshift(bits, num_positions):
    shifted = []
    for i in range(num_positions, len(bits)):
        shifted.append(bits[i])
    for i in range(num_positions):
        shifted.append(bits[i])
    return shifted

#XOR together
def xor(bits_one, bits_two):
    output = []
    for i in range(len(bits_one)):
        output.append(bits_one[i] ^ bits_two[i])
    return output
```

```python
def generate_16_round_keys(keyBits):
    #PC1 = drop every 8th bit and permute and rearrange into 56 bits
    keyBits56 = permute(keyBits, PC_1)

    #split into two halfs
    C = keyBits56[:28]
    D = keyBits56[28:]

    roundKeys = []
    for i in range(16):
        #shift C left
        C = Lshift(C, LeftShifts[i])
        #shift D left
        D = Lshift(D, LeftShifts[i])
        #PC2 = drop 8 more bits and rearrange into 48 bit subkey Kn
        Kn = permute(C + D, PC_2)
        roundKeys.append(Kn)
    return roundKeys
```

```python
def des_decryption(cipherBits, roundKeys):
    ip_result = permute(cipherBits, IP)
    L = ip_result[:32]
    R = ip_result[32:]

    for n in range(16):
        #rounds in reverse 16 to 1
        K = roundKeys[15 - n]
        #expand R 32 to 48 bits
        Expanded_R = permute(R, E_bit_selection_table)
        #XOR with round key
        xored = xor(Expanded_R, K)
        result  = []
        #feed 6 bit parts into 8 S-boxes to get 4 bits
        for i in range(8):
            part = xored[i*6:(i+1)*6]
            #add first and last bit for row
            row = int(''.join(map(str, [part[0], part[5]])), 2)
            #middle 4 bits are col
            col = int(''.join(map(str, part[1:5])), 2)
            #4 bit output using right shift
            output = Sbox_X[i][row][col]
            for j in range(4):
                bit = (output >> (3 - j)) & 1
                result.append(bit)
        #P permutation
        f = permute(result, P)
        #swap and XOR
        new_L = R
        new_R = xor(L, f)
        L = new_L
        R = new_R

        #print details
        print("Iteration " + str(n+1) + ":")
        print("K" + str(16-n) + ": " + ''.join(map(str, K)))
        print("f: " + ''.join(map(str, f)))
        print("L: " + ''.join(map(str, L)))
        print("R: " + ''.join(map(str, R)))
        print("\n")

    # swap L and R
    swap = R + L

    #apply inverse IP
    plainBits = permute(swap, inverseIP)  # final swap + inverse IP
    return plainBits
```
